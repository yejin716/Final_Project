전처리

```python
import pandas as pd
import numpy as np
import re
# 정규 표현식 함수 정의

def apply_regular_expression(text):
    hangul = re.compile('[^ㄱ-ㅎ가-힣0-9a-zA-Z\s]')  # 한글 추출 규칙: 띄어 쓰기(1 개)를 포함한 한글
    result = hangul.sub('', text)  # 위에 설정한 "hangul"규칙을 "text"에 적용(.sub)시킴
    return result

df['reviews'] = df['reviews'].apply(lambda x: apply_regular_expression(x))
df['reviews']
```

https://huggingface.co/klue/bert-base

```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline

# 리뷰 데이터 로드

# 사전 학습된 모델과 토크나이저 로드
model_name = 'klue/bert-base'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=2)  # 3가지 레이블 (긍정, 부정)

# 감정 분석 파이프라인 생성
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# 데이터에 감성 분석 적용
def analyze_sentiment(review):
    result = nlp(review[:512])  # BERT 모델의 입력 제한 고려 (최대 512 토큰)
    return result[0]['label']

df['sentiment'] = df['reviews'].apply(analyze_sentiment)

# 결과 출력
print(df.head())
```
