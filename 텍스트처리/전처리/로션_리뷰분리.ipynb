```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = '/content/drive/MyDrive/최종프로젝트/data/Lotion_dataset.csv'

raw = pd.read_csv(path)
df = raw.copy()
```

```python
# 중복되는 이름 확인
duplicate_names = df[df.duplicated('name', keep=False)]

duplicate_names.head(3)
```

```python
#중복 제거
df.drop_duplicates()[:3]
```

```python
lotion = df.drop_duplicates(subset = 'name')
lotion.shape
```

```python
lotion['reviews'][0]
```

```python
reviews = lotion['reviews']

lotion['reviews'] = reviews.str.replace("'","").str.replace('"',"")
lotion['reviews'] = reviews.str.replace("순해서 너무 좋으나,","순해서 너무 좋으나").str.replace("'","")
# str.replace("[","").str.replace("]","")
# 결과 출력
lotion['reviews']
```

```python
lotion.groupby('name')['reviews'].count()
```

```python
sentence_list = []
for review in lotion['reviews'][:80]:
    sentences = review.split(',')
    sentence_list.extend(sentences)
```

```python
lotion['reviews'] = sentence_list
```
