```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = '/content/drive/MyDrive/최종프로젝트/data/Lotion_dataset.csv'

raw = pd.read_csv(path)
df = raw.copy()
```

```python
# 중복되는 이름 확인
duplicate_names = df[df.duplicated('name', keep=False)]

duplicate_names.head(3)
```

```python
lotion = df.drop_duplicates(subset = 'name')
lotion.shape
```

```python
lotion[['reviews']][:1]
```

```python
# 'review' 열에서 '], ['를 기준으로 문자열 분리
def split_reviews(review_str):
    # 문자열에서 '], ['를 기준으로 분리한 후, 앞뒤의 대괄호 제거
    if isinstance(review_str, str):
        # '], ['를 기준으로 분리
        parts = review_str.split('", "')
        # 대괄호 제거
        parts = [part.strip('[]"').strip("'") for part in parts]
        return parts
    else:
        return []
# 'review' 열에 함수 적용
lotion['reviews'] = lotion['reviews'].apply(split_reviews)
```

```python
lotion.reviews[1][0]
```

```python
# 각 리뷰를 별도의 행으로 변환하는 함수 정의
def explode_reviews(df):
    # DataFrame의 각 행에 대해 'review' 열을 열의 요소로 분리하여 새로운 행 생성
    return df.explode('reviews')

# 변환 적용
expanded_lotion_data = explode_reviews(lotion)
expanded_lotion_data
```

```python
#제품별 리뷰 개수
lotion_count = expanded_lotion_data.groupby('name').size().reset_index(name='count')
lotion_count
```

```python
lotion_count[lotion_count['count'] < 80]
```

```python
expanded_lotion_data.shape
```

```python
expanded_lotion_data = expanded_lotion_data.drop_duplicates(subset=('name', 'reviews'))
expanded_lotion_data.shape
```

```python
lotion_count[lotion_count['count'] < 80].name
```

```python
#리뷰 80개 미만 제거
condition = lotion[lotion.name.isin(lotion_count[lotion_count['count'] < 80].name)]
condition
```

```python
lotion = lotion[lotion['name'] != '비욘드 엔젤아쿠아 보습 장벽 로션 200ml']
```

```python
lotion.shape
```

```python
def collapse_review(df): # 행으로 변환한 리뷰를 다시 하나의 리스트 형식으로
  return df.groupby('name', as_index = False).agg({'reviews': lambda x : list(x)})

lotion_data = collapse_review(expanded_lotion_data)
lotion_data.shape
```

```python
# 각 제품의 리뷰 수를 80개로 맞추는 함수 정의
def sample_reviews(reviews, n=80):
    if len(reviews) > n:
        return pd.Series(reviews).sample(n=n, random_state=1).tolist()
    else:
        return reviews
```

```python
lotion_data['reviews'] = lotion_data['reviews'].apply(sample_reviews)
```

```python
lotion_data.columns = ['name', 'sample_review']
```
